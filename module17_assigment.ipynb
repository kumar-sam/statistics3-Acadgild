```python
# Assignment 17.1
"""
Question:
Blood glucose levels for obese patients have a mean of 100 with a standard deviation of
15. A researcher thinks that a diet high in raw cornstarch will have a positive effect on
blood glucose levels. A sample of 36 patients who have tried the raw cornstarch diet
have a mean glucose level of 108. Test the hypothesis that the raw cornstarch had an
effect or not.
"""
"""
Answer:
The above question is an example of two tailed test.
Here we will assume the null hypothesis(H0) and alternative hypothesis(H1)

(population mean) mu=100 
(population standard deviation) sigma=15

(a random sample mean) x_bar=108
(size of the random sample) n=36

since significance level is not given, hence we will assume significance level equal to 95%
hence, alpha=0.25

In standard normal distribution value of Z_alpha_divide_2 will be 
Z_alpha_divide_2= 1.96

H0:mu=100
h1:mu!=100 (mu>100 | mu<100)

we will find Z value for sample mean (Z_x_bar)

Z_x_bar =(x_bar -mu)/(sigma/sqrt(n))
        =(108-100)/(15/sqrt(36))
        =8/2.5
        =3.2

we can observe that  Z_x_bar>>Z_alpha_divide_2 lies in the rejection region.
hence we reject the null hypothesis.
so we accept the alternative hypothesis(H1)

         Hence,with 95% confidence interval we can say that raw cornstarch had an effect on glucose level.

"""
```
